```python
# Define survey responses data
survey_data = {
    "CoordinationSatisfaction": [1, 1, 1, 1, 1, 1, 2],
    "StructuresSatisfaction": [1, 1, 2, 1, 1, 2, 1],
    "ListingTeamSatisfaction": [1, 2, 2, 2, 2, 2, 1],
    "SurveyTeamSatisfaction": [1, 1, 1, 1, 1, 1, 3],
    "OfficersSatisfaction": [1, 1, 1, 1, 1, 1, 1],
}
```

```python
# Define categories for satisfaction levels (map survey values to labels)
satisfaction_levels = {1: "To a large extent", 2: "To a limited extent", 3: "Not followed at all"}
```

```python
# Create subplots for each satisfaction category
fig, axes = plt.subplots(2, 3, figsize=(16, 8))

# Iterate through survey data and plot satisfaction distribution for each category
for i, (category, responses) in enumerate(survey_data.items()):
    row, col = divmod(i, 3)  # Calculate row and column index for subplot
    axes[row, col].hist(responses, bins=3, edgecolor='black')
    axes[row, col].set_title(category)
    axes[row, col].set_xlabel("Satisfaction Level")
    axes[row, col].set_ylabel("Frequency")
    axes[row, col].set_xticks([1, 2, 3])
    axes[row, col].set_xticklabels(satisfaction_levels.values())

# Tight layout to prevent overlapping labels
plt.tight_layout()

# Display the satisfaction graph
plt.show()
```

```python
# Create a new figure
plt.figure(figsize=(12, 6))  # Adjust figure size for horizontal layout

# Create horizontal stacked bar chart
bar_height = 0.5  # Adjust bar height as needed
index = range(len(survey_data))  # Create y-axis positions

# Initialize left values for stacking
left = [0] * len(survey_data)

# Iterate through satisfaction levels and stack the bars
for level, label in satisfaction_levels.items():
    satisfaction_counts = [data.count(level) for data in survey_data.values()]
    plt.barh(index, satisfaction_counts, bar_height, left=left, label=label)
    left = [left[i] + satisfaction_counts[i] for i in range(len(satisfaction_counts))]

# Set labels and title
plt.ylabel("Categories")
plt.xlabel("Number of Respondents")
plt.title("Satisfaction Levels by Category for Enumerators")

# Set y-axis positions and labels for categories
plt.yticks(index, survey_data.keys())

# Add legend
plt.legend()

# Show the chart
plt.tight_layout()
plt.show()
```
